## Exercício 1

```python
def exercicio_1(n):
    lista = []
    for x in range (1,n+1):
        lista += [(x,x**2)]
    return lista
exercicio_1(3)
```

## Exercício 2

```python
def exercicio_2(n):
    lista = []
    for x in range (1,n+1):
        lista += [(x,x**2,x**3)]
    return lista
exercicio_2(3)
```

## Exercicio_7
