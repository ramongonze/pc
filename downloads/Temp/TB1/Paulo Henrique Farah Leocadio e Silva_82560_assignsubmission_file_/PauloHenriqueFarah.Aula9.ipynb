```python
def exercicio9_1(n):
    lista = []
    for i in range(1,n+1):
        tupla = (n, n**2)
        n=n-1
        lista.append(tupla)
    return lista[::-1]
exercicio9_1(5)
    
```

```python
def exercicio9_2(n):
    lista = []
    for i in range(1,n+1):
        tupla = (n, n**2,n**3)
        n=n-1
        lista.append(tupla)
    return lista[::-1]
exercicio9_2(5)
```
