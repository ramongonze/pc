```python
#n=int(input('Digite o valor de n: '))
def exercicio_1(n):
#função que retorna uma lista de tupla com os valores inteiros e seus quadrados de 1 até n
    a=[]
    i=1
    while i<=n:
        a.append((i,i**2))
        i+=1
    return a
#exercício_9_1(n)
```

```python
#n=int(input('Digite o valor de n: '))
def exercicio_2(n):
#função que retorna uma lista de tupla com os valores inteiros, seus quadrados e cubos de 1 até n
    a=[]
    i=1
    while i<=n:
        a.append((i,i**2,i**3))
        i+=1
    return a
#exercício_9_2(n)
```

```python
def exercicio_3(a):
    b=()
    c=()
    m=len(a)
    for e in range (m//2):
        b+=a[e],
    for f in range ((len(a)//2),len(a)):
        c+=a[f],
    return(str(b)+"\n"+str(c))
#print(exercício_9_3(a))
```
