```python
#EXERCICIO 1

def exercicio_1(n):
    lista=[]
    for i in range(1,n+1):
        exercicio_1 = [i,i*i]
        lista.append(exercicio_1)
    return lista

print (exercicio_1(10))

 
    
    
```

```python
#EXERCICIO 2

def exercicio_2(n):
    lista=[]
    for i in range(1,n+1):
        exercicio_2 = [i,i*i,i*i*i]
        lista.append(exercicio_2)
    return lista

print (exercicio_2(10))

 
```

```python
#EXERCICIO 4

def exercicio_4(t):
    par, impar = (), ()
    for i in t:
        if i%2 == 0:
            par += (i,)
        else:
            impar += (i,)

    return par, impar

print(exercicio_4((1,2,3,4,5)))
```

```python
#Exercicio 5

def exercicio_5_1(l):
    return tuple(l)

def exercicio_5_2(l):
    tupla = ()
    for i in l:
        tupla += (i,)
    return tupla

print(exercicio_5_1([1, 4]))
print(exercicio_5_2([1, 4]))
```

```python
#exercicio 6

lista=[(10,20,40), (40, 50, 60), (70, 80, 90)]

def exercicio_6(l,n):
    lista=[]
    for tupla in l:
        novaTupla = tupla[:-1] + (n,)
        lista.append(novaTupla)
    return lista

print(exercicio_6(lista, 100))
    
```

```python
#Exercicio 7

def exercicio_7(cotacao):
    newList = []
    for i in range(1, 101):

        tuplaDR = (i, i*cotacao)
        tuplaRD = (i*cotacao, i)

        tupla = (tuplaRD, tuplaDR)
        newList.append(tupla)

    return newList

print(exercicio_7(3.98))
```
