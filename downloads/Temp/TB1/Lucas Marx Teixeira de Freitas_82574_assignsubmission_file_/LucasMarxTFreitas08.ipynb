```python
def exercicio_4(t):
    length = len(t)
    sort = sorted(t)
    x = ()
    y = ()
    z = ()
    w = ()
    for b in range(0,length):
        if sort[b] %2 == 0:
            z += sort[b],
            x = z
        if sort[b] %2 != 0:
            w += sort[b],
            y = w
    return(x,y)
            
            
```
