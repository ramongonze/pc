```python
def pares( tupla ):

    listone = []

    for k in tupla:
        if k % 2 == 0:
            listone.append(k)

    return listone


tple = (1,2,3,214,35456,567,54)
one = pares( tple )

print(one)
```

```python
def impares( tupla ):

    listtwo = []

    for k in tupla:
        if k % 2 != 0:
            listtwo.append(k)

    return listtwo


tple = (1,2,3,214,35456,567,54)
two = impares( tple )

print(two)
```

```python
cotacaodolar = float(input("Digite a cotação atual do dolar: "))

def dolarreal (dolar):
    real = dolar*cotacaodolar
    dicionariodolarreal = sorted({("USD = %.2f"%(dolar)),("BRL = %.2f"%(real))})
    dicionariodolarreal.sort (reverse = True)
    return dicionariodolarreal

print ("\nUSD para BRL:")
print ("-----------------------")

for dolar in range (1,101):
    string = str (dolarreal(dolar))
    print (string.replace("[","").replace("]","").replace(",","").replace("(","").replace(")","").replace("'",""))
    print ("-----------------------")
    
print ("\nBRL para USD:")
print ("-----------------------")
 
def realdolar (real):
    dolar = real/cotacaodolar
    dicionariodolarreal = sorted({("BRL = %.2f"%(real)),("USD = %.2f"%(dolar))})
    return dicionariodolarreal

for real in range (1,101):
    string = str (realdolar(real))
    print (string.replace("[","").replace("]","").replace(",","").replace("(","").replace(")","").replace("'",""))
    print ("-----------------------")
```
