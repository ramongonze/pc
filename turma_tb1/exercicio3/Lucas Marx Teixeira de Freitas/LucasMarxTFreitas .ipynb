As duas operações somente são semelhantes quando: 
x = True e y = False e quando 
x = False e y = True

```python
#Tabela da operação lógica ↔
print(" x       y    x  ↔ y ") #(a ↔ b) = A → B ∧ B → A. = not x or y and not y or x
x = True; y = True
print(x," ",y," ",   not x or y and not y or x) 
x = True; y = False
print(x," ",y," ",   not x or y and not y or x)    
x = False ; y = True
print(x," ",y," ",   not x or y and not y or x) 
x = False ; y = False
print(x," ",y," ",   not x or y and not y or x)
```
